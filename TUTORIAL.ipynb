**RESPRESENTAR UNA MATRIZ**

```python
# Definimos una matriz de 2x2
A= sp.Matrix([[1, 2], [3, 4]])

# Imprimimos la matriz
print("Matriz A:")
sp.pprint(A)
```

***PARA OBTENER ENTRADAS ESPECIFICAS***

```python
renglon_1 = A.row(0)   #Primera fila
columna_2 = A.col(1)   #Segunda columna
entrada_especifica = A[1, 0]  #Fila 2
#Columna 1 (el numero 3)
```

```python
print(f"\nEntrada en [1, 0]:\n    {entrada_especifica}")
```

**MULTIPLICACION DE MATRICES**

```python
B = sp.Matrix([[5, 6],
              [7, 8]])
producto = A * B
print("\nA * B:")
sp.pprint(producto)
```

**ELEVAR A UNA POTENCIA AL CUADRADO $= A^2$**

```python
al_cuadrado = A**2
print("\nA elevada al cuadrado:")
sp.pprint(al_cuadrado)
```

**MULTIPLICAR VECTORES POR MATRICES**

```python
v = sp.Matrix([1, 2])   #Vector columna
resultado = A * v
print("\nA * v:")
sp.pprint(resultado)
```

**VALORES Y VECTORES PROPIOS**

```python
# .eigenvals() devuelve {valor: multiplicidad}
# .eigenvects() devuelve [(valor, multiplicidad, [vector]), ...]


print("\nValores propios:",
      A.eigenvals())

print("Vectores propios:")
sp.pprint(A.eigenvects())
```

**DIAGRAMA DE TRANSICION**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Crear un grafico dirigido
G = nx.DiGraph()
```

**Definir los estados y sus probabilidades de transicion**

```python
# Formato: (Origen, Destino, Probabilidad)
transiciones = [
("Estado A", "Estado A", 0.3),
("Estado A", "Estado B", 0.7),
("Estado B", "Estado A", 0.4),
("Estado B", "Estado B", 0.6)
]
```

**Añadiendo las aristas con sus pesos**

```python
for u, v, w in transiciones:
    G.add_edge(u, v, weight=w)
```

**Finalmente:**

```python
# Dibujar el diagrama
plt.figure(figsize=(8, 5))
pos = nx.spring_layout(G) # Posicionamiento de los nodos

# Dibujar nodos y etiquetas
nx.draw(G, pos, with_labels=True, node_color='yellow', node_size=2000, font_weight='bold', arrowsize=20)

# Dibujar las etiquetas de las probabilidades sobre las flechas
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.title("Diagrama de Transición")
plt.show()
```
