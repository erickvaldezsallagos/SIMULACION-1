**CONTEO DE RACHAS**

```python
# Si Xi <= Xi-1 -> escribir 0
# Si Xi > Xi-1 -> escribir 1
secuencia_s = []

for i in range(1, len(lista_datos)):
    if lista_datos[i] <= lista_datos[i-1]:
        secuencia_s.append(0)
    else:
        secuencia_s.append(1)
```

```python
# Una racha es una secuencia de ceros o unos consecutivos
if not secuencia_s:
    co = 0
else:
    co = 1 # Iniciamos en 1 por la primera racha existente
    for i in range(1, len(secuencia_s)):
        if secuencia_s[i] != secuencia_s[i-1]:
            co += 1
```
