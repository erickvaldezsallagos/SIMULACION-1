**Calculamos los estadisticos**

### Pruebas de Bondad de Ajuste: Kolmogorov-Smirnov
La prueba busca determinar si la muestra $X_n$ sigue una distribución uniforme $U(0,1)$.

*Hipótesis:*
* $H_0: x_i \sim U(0,1)$
* $H_a: x_i \nsim U(0,1)$

*Fórmulas de los estadísticos:*
Para el cálculo, primero ordenamos los datos de menor a mayor y luego aplicamos:

1. *Estadístico superior ($D^+$):*
$$D^+ = \max_{1 \leq i \leq n} \left\{ \frac{i}{n} - x_i \right\}$$

2. *Estadístico inferior ($D^-$):*
$$D^- = \max_{1 \leq i \leq n} \left\{ x_i - \frac{i-1}{n} \right\}$$

3. *Estadístico de prueba ($D$):*
$$D = \max(D^+, D^-)$$

```python
# Creamos listas para almacenar los cálculos de cada columna
i_list = list(range(1, n + 1))
xi = datos_ordenados
i_n = [i/n for i in i_list]
i_minus_1_n = [(i-1)/n for i in i_list]
d_mas = [abs((i/n) - x) for i, x in zip(i_list, xi)]
d_menos = [abs(x - ((i-1)/n)) for i, x in zip(i_list, xi)]
```

```python
# Creamos el DataFrame para mostrar la tabla igual que en tu libreta
df = pd.DataFrame({
    'i': i_list,
    'xi': xi,
    '(i-1)/n': i_minus_1_n,
    'i/n': i_n,
    'D-': d_menos,
    'D+': d_mas
})
```

```python
# Calculamos el estadístico final D
D_max_mas = max(d_mas)
D_max_menos = max(d_menos)
D_final = max(D_max_mas, D_max_menos)

print(f"Estadístico D+ máximo: {D_max_mas:.5f}")
print(f"Estadístico D- máximo: {D_max_menos:.5f}")
print(f"Valor calculado de D: {D_final:.5f}")
```

**COMPARACION CON EL VALOR CRITICO**

**CONCLUSION**

```python

if D_final < valor_critico:
    print(f"Como {D_final:.5f} < {valor_critico:.5f}, NO SE RECHAZA H0.")
    print("Conclusión: Los datos siguen una distribución uniforme.")
else:
    print(f"Como {D_final:.5f} > {valor_critico:.5f}, SE RECHAZA H0.")
    print("Conclusión: Los datos NO siguen una distribución uniforme.")
```
