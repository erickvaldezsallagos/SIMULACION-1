**Solucion Analitica:**

**a)**
Sea:


$X_1 \sim N(1.5,\;0.0016)$



$X_2 \sim N(1.48,\;0.0009)$


Hay interferencia cuando la flecha es más grande que el cojinete:

$X_2 > X_1$


Definimos la variable:

$Y = X_2 - X_1$

Como ambas variables son normales independientes:

$Y \sim N(\mu_Y,\sigma_Y^2)$

con:
$\mu_Y = 1.48 - 1.50 = -0.02$

$\sigma_Y^2 = 0.0009 + 0.0016 = 0.0025$

$\sigma_Y = \sqrt{0.0025}=0.05$

Entonces:
$Y \sim N(-0.02,0.0025)$

La probabilidad de interferencia es:

$P(X_2>X_1)=P(Y>0)$

Estandarizando:
$Z=\frac{0-(-0.02)}{0.05}$

$Z=0.4$

Entonces:

$P(Y>0)=P(Z>0.4)$
Usando la tabla normal estándar:

$P(Z>0.4)=1-0.6554$

$P(Z>0.4)=0.3446$

$\boxed{P(\text{interferencia})\approx 0.3446}$






**b)**

Número de Simulaciones Necesarias

Se desea que:


$| \hat p - p | < 0.01$

con un nivel de confianza del $95$%.

Usamos la fórmula:

$n=\frac{z_{\alpha/2}^2p(1-p)}{E^2}$

Sustituyendo:

$n=\frac{(1.96)^2(0.3446)(1-0.3446)}{(0.01)^2}$

$n=\frac{3.8416(0.3446)(0.6554)}{0.0001}$

$n\approx 8676.28$

Resultado

$\boxed{n \approx 8677 \text{ simulaciones}}$

**Solucion (Por Simulacion):**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**De a)**

**Parametros:**

```python
mu1 = 1.5
var1 = 0.0016
sigma1 = np.sqrt(var1)

mu2 = 1.48
var2 = 0.0009
sigma2 = np.sqrt(var2)
```

De lo anteorior, tomamos el tamño de simulaciones:

```python
n = 100000

# Simulacion
x1 = np.random.normal(mu1, sigma1, n)
x2 = np.random.normal(mu2, sigma2, n)
```

INTERFERENCIA:

```python
interferencia = x2 > x1
```

**Probabilidad:**

```python
p_estimada = np.mean(interferencia)
print("Probabilidad estimada:", p_estimada)
```

**Conclusion:**

La solución por simulación coincide con la solución analítica. El modelo probabilístico utilizado es correcto y que el método Monte Carlo permite aproximar adecuadamente la probabilidad de interferencia del ensamblaje.
