# **Generacion de una variable de Poisson**

**Interpretacion:**

La distribución de Poisson es una distribución de probabilidad discreta que modela el número de veces que ocurre un evento en un intervalo determinado, suponiendo que los eventos suceden de forma independiente y con una tasa promedio constante $\lambda$.

Esta distribución tiene gran utilidad en simulación y procesos estocásticos porque permite representar fenómenos reales donde se cuentan eventos discretos. Algunos ejemplos son: el número de llamadas que recibe un centro de atención por hora, la cantidad de clientes que llegan a una tienda, el número de fallas de una máquina durante un día o la cantidad de accidentes en una intersección.

En este ejercicio se utiliza el método de la transformada inversa para generar valores aleatorios con distribución Poisson. Este método consiste en generar números aleatorios uniformes y transformarlos mediante la función de distribución acumulada hasta obtener observaciones que reproduzcan el comportamiento probabilístico de la distribución teórica. Esto permite realizar simulaciones y analizar el comportamiento esperado del fenómeno modelado.

**Tomando en cuenta el Seucodigo:**

```python
import random
import math
import matplotlib.pyplot as plt
from collections import Counter
```

**Parametros de Poisson**

```python
lam = 4

# Simulaciones
N = 10000

# Lista para guardar los resultados
muestras = []
```

**Ahora, realizamos la simulacion por Transformada Inversa**

```python
for _ in range(N):

    U = random.uniform(0, 1)

    i = 0
    p = math.exp(-lam)   # p0
    F = p

    while U >= F:
        i += 1
        p = (lam * p) / i
        F += p

    muestras.append(i)
```

**Frecuencias relativas:**

```python
conteo = Counter(muestras)

print("Probabilidades simuladas:")

for k in sorted(conteo):
    print(f"P(X={k}) ≈ {conteo[k]/N:.4f}")

```

**Histograma:**

```python

plt.figure(figsize=(8,5))

plt.hist(
    muestras,
    bins=range(min(muestras), max(muestras)+2),
    density=True
)

plt.xlabel("Valores de X")
plt.ylabel("Frecuencia relativa")
plt.title(f"Simulación Poisson λ = {lam}")

plt.show()
```

**Conclusion:**

El código permitió simular correctamente una distribución de Poisson usando el método de Transformada Inversa. Los resultados y el histograma obtenidos se aproximan a la distribución teórica con $λ=4$, mostrando que la simulación representa adecuadamente el comportamiento esperado de esta distribución de probabilidad.
