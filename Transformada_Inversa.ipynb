```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

La variable aleatoria $X$ se define como:
$$X = \cases{\sqrt{2R}&si $0 \leq x \leq \frac{1}{2}$,\cr 2-\sqrt{2(1-R)}&si $\frac{1}{2} < R \leq 1$}$$

```python
def triangular(n):
    l = []
    for i in range(n):
        u = r.random()
        if 0 <= u < 0.5:
            x = np.sqrt(2*u)
            l.append(x)
        else:
            x = 2-np.sqrt(2*(1-u))
            l.append(x)
    return l
```

```python
M = triangular(10000)
```

```python
plt.hist(M, bins=50)
```

```python
def cdf(M):
    n=len(M)
    x=np.sort(M)
    y=np.arange(0,n)/n
    return x,y
```

```python
x_math,y_math=cdf(M)
```

```python
media = np.mean(M)
```

```python
plt.step(x_math,y_math)
plt.axhline(y=0.5, color='red') #La mediana
plt.axvline(x=media, color='green') # La media
```

Resolver los siguientes problemas usando la transformada inversa de las siguientes distribuciones acumuladas:

a) *Distribución de Weibull:* $\mathcal{We}(\alpha,1)$

b) *Distribución de Gumbel estándar:* $\mathcal{Gu}(0,1)$

c) *Distribución de Cauchy estándar:* $\mathcal{C}(0,1)$

d) *Distribución Logística estándar:* $\mathcal{L}(0,1)$

e) *Distribución de Pareto (con $xm=1$):* $\mathcal{Par}(\alpha)$

**Solución:**

a) *Distribución de Weibull:* $\mathcal{We}(\alpha,1)$

Función acumulativa: $F(x)=1-e^{-x^\alpha}, x \geq 0$

Variable aleatoria $X$: $x=(-\ln U)^{\frac{1}{\alpha}}$

```python
def weibull(n,alpha=5):
    l = []
    for i in range(n):
        u = r.random()
        if u >= 0:
            x = (-np.log(u))**(1/alpha)
            l.append(x)
        else:
            x = 0
            l.append(x)
    return l
```

```python
W = weibull(10000)
```

```python
plt.hist(W, bins=50)
```

```python
def cdfa(W):
    n=len(W)
    x=np.sort(W)
    y=np.arange(0,n)/n
    return x,y
```

```python
xa_math,ya_math=cdfa(W)
```

```python
mediaa = np.mean(W)
```

```python
plt.step(xa_math,ya_math)
plt.axhline(y=0.5, color='red') #La mediana
plt.axvline(x=mediaa, color='green') # La media
```

b) *Distribución de Gumbel estándar:* $\mathcal{Gu}(0,1)$

Función acumulativa: $F(x)=e^{-e^{-x}}$

Variable aleatoria $X$: $x=-ln(-\ln U)$

```python
def gumbel(n):
    l = []
    for i in range(n):
        u = r.random()
        if u > 0:
            x = -np.log(-np.log(u))
            l.append(x)
        else:
            x = 0
            l.append(x)
    return l
```

```python
G = gumbel(10000)
```

```python
plt.hist(G, bins=50)
plt.title("Histograma Distribución Gumbel")
plt.show()
```

c) *Distribución de Cauchy estándar:* $\mathcal{C}(0,1)$

Función acumulativa: $F(x)=\frac{1}{\pi}\arctan(x)+\frac{1}{2}$

Variable aleatoria $X$: $x=\tan [\pi\left(U-\frac{1}{2}\right)]$

```python
def cauchy(n):
    l = []
    for i in range(n):
        u = r.random()
        x = np.tan(np.pi*(u - 0.5))
        l.append(x)
    return l
```

```python
C = cauchy(10000)
```

```python
plt.hist(C, bins=50, range=(-10, 10))
plt.title("Histograma Distribución Cauchy (Rango limitado)")
plt.show()
```

d) *Distribución Logística estándar:* $\mathcal{L}(0,1)$

Función acumulativa: $F(x)=\frac{1}{1+e^{-x}}$

Variable aleatoria $X$: $x=\ln \left(\frac{U}{1-U}\right)$

```python
def logistica(n):
    l = []
    for i in range(n):
        u = r.random()
        if u > 0 and u < 1:
            x = np.log(u/(1-u))
            l.append(x)
        else:
            x = 0
            l.append(x)
    return l
```

```python
L = logistica(10000)
```

```python
plt.hist(L, bins=50)
plt.title("Histograma Distribución Logística")
plt.show()
```

e) *Distribución de Pareto (con $xm=1$):* $\mathcal{Par}(\alpha)$

Función acumulativa: $F(x)=1-x^{-\alpha}, x \geq 1$

Variable aleatoria $X$: $x=(1-U)^{-\frac{1}{\alpha}}$

Equivalente: $x=U^{-\frac{1}{\alpha}}$

```python
def pareto(n, alpha=3):
    lista = []
    for i in range(n):
        u = r.random()
        # Se aplica la transformada inversa: x = u**(-1/alpha)
        x = u**(-1/alpha)
        lista.append(x)
    return lista
```

```python
P = pareto(10000, alpha=2.5)
```

```python
# Histograma normalizado
count, bins, ignored = plt.hist(P, bins=50, density=True)

# Función de densidad teórica de Pareto
x = np.linspace(1, max(P), 1000)
alpha = 2
f = alpha * x**(-(alpha+1))

# Graficar función
plt.plot(x, f)
plt.xlim(1, 10)
plt.show()
```
