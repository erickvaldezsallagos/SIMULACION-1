```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

La variable aleatoria $X$ se define como:
$$X = \cases{\sqrt{2R}&si $0 \leq x \leq \frac{1}{2}$,\cr 2-\sqrt{2(1-R)}&si $\frac{1}{2} < R \leq 1$}$$

```python
def triangular(n):
    l = []
    for i in range(n):
        u = r.random()
        if 0 <= u < 0.5:
            x = np.sqrt(2*u)
            l.append(x)
        else:
            x = 2-np.sqrt(2*(1-u))
            l.append(x)
    return l
```

```python
M = triangular(10000)
```

```python
plt.hist(M, bins=50)
```

```python
def cdf(M):
    n=len(M)
    x=np.sort(M)
    y=np.arange(0,n)/n
    return x,y
```

```python
x_math,y_math=cdf(M)
```

```python
media = np.mean(M)
```

```python
plt.step(x_math,y_math)
plt.axhline(y=0.5, color='red') #La mediana
plt.axvline(x=media, color='green') # La media
```

Resolver los siguientes problemas usando la transformada inversa de las siguientes distribuciones acumuladas:

a) *Distribución de Weibull:* $\mathcal{We}(\alpha,1)$

b) *Distribución de Gumbel estándar:* $\mathcal{Gu}(0,1)$

c) *Distribución de Cauchy estándar:* $\mathcal{C}(0,1)$

d) *Distribución Logística estándar:* $\mathcal{L}(0,1)$

e) *Distribución de Pareto (con $xm=1$):* $\mathcal{Par}(\alpha)$

**Solución:**

a) *Distribución de Weibull:* $\mathcal{We}(\alpha,1)$

Función acumulativa: $F(x)=1-e^{-x^\alpha}, x \geq 0$

Variable aleatoria $X$: $x=(-\ln U)^{\frac{1}{\alpha}}$

```python
def weibull(n,alpha=5):
    l = []
    for i in range(n):
        u = r.random()
        if u >= 0:
            x = (-np.log(u))**(1/alpha)
            l.append(x)
        else:
            x = 0
            l.append(x)
    return l
```

```python
W = weibull(10000)
```

```python
plt.hist(W, bins=50)
```

```python
def cdfa(W):
    n=len(W)
    x=np.sort(W)
    y=np.arange(0,n)/n
    return x,y
```

```python
xa_math,ya_math=cdfa(W)
```

```python
mediaa = np.mean(W)
```

```python
plt.step(xa_math,ya_math)
plt.axhline(y=0.5, color='red') #La mediana
plt.axvline(x=mediaa, color='green') # La media
```

b) *Distribución de Gumbel estándar:* $\mathcal{Gu}(0,1)$

Función acumulativa: $F(x)=e^{-e^{-x}}$

Variable aleatoria $X$: $x=-ln(-\ln U)$

c) *Distribución de Cauchy estándar:* $\mathcal{C}(0,1)$

Función acumulativa: $F(x)=\frac{1}{\pi}\arctan(x)+\frac{1}{2}$

Variable aleatoria $X$: $x=\tan [\pi\left(U-\frac{1}{2}\right)]$

d) *Distribución Logística estándar:* $\mathcal{L}(0,1)$

Función acumulativa: $F(x)=\frac{1}{1+e^{-x}}$

Variable aleatoria $X$: $x=\ln \left(\frac{U}{1-U}\right)$

e) *Distribución de Pareto (con $xm=1$):* $\mathcal{Par}(\alpha)$

Función acumulativa: $F(x)=1-x^{-\alpha}, x \geq 1$

Variable aleatoria $X$: $x=(1-U)^{-\frac{1}{\alpha}}$

Equivalente: $x=U^{-\frac{1}{\alpha}}$
