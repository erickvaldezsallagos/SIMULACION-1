# **LINEA DE ESPERA CON UN SERVIDOR**

**Tomando en cuenta el Seucodigo de Ross, tomemos en cuenta:**
$λ=4$ , $μ= 6$

Resolviendo analiticamente

Datos:

$\lambda = 4$\
$\mu = 6$

Como ($\lambda < \mu$), el sistema se encuentra en estado estable.

**Factor de utilizacion:**

$\rho = \frac{\lambda}{\mu}$


Sustituyendo:
$\rho = \frac{4}{6}= 0.666667$


**Probabilidad de que no haya clientes en cola:**


$P_0 = 1-\rho$
$P_0 = 1-0.666667=0.333333$


**Número promedio de clientes en cola**


$L_q = \frac{\lambda^2}{\mu(\mu-\lambda)}$

Sustituyendo:\
$L_q =\frac{4^2}{6(6-4)}=\frac{16}{12}=1.333333$

**Numero promedio de clientes en el sistema:**

$L = L_q + \frac{\lambda}{\mu}$\
$L = 1.333333 + \frac{4}{6}$\
$L = 1.333333 + 0.666667$\
$L = 2$

**Tiempo promedio de espera en cola:**

$W_q = \frac{L_q}{\lambda}$\
$W_q =\frac{1.333333}{4}$\
$W_q = 0.333333$

**Tiempo promedio en el sistema:**\
$W = W_q + \frac{1}{\mu}$\
$W = 0.333333 + \frac{1}{6}$\
$W = 0.333333 + 0.166667$\
$W = 0.500000$

**Solucion (Por Simulacion):**

```python
import numpy as np
```

Definimos la funcion de Ross

```python
def simulacion_ross_m_m_1(lam, mu, T):
    t, Na, Nd, n = 0.0, 0, 0, 0
    t_A = np.random.exponential(1 / lam)
    t_D = float('inf')

    A, D = {}, {}

    while True:
        if min(t_A, t_D) > T and n == 0:
            Tp = max(t - T, 0)
            break

        if t_A <= t_D and t_A <= T:
            t = t_A
            Na += 1
            n += 1
            t_A = t + np.random.exponential(1 / lam)
            if n == 1:
                t_D = t + np.random.exponential(1 / mu)
            A[Na] = t

        elif t_D < t_A and t_D <= T:
            t = t_D
            n -= 1
            Nd += 1
            D[Nd] = t
            t_D = float('inf') if n == 0 else t + np.random.exponential(1 / mu)

        elif min(t_A, t_D) > T and n > 0:
            t = t_D
            n -= 1
            Nd += 1
            D[Nd] = t
            if n > 0:
                t_D = t + np.random.exponential(1 / mu)

    clientes_completados = min(Na, Nd)
    tiempos_en_sistema = [D[i] - A[i] for i in range(1, clientes_completados + 1)]
    Ws_promedio = np.mean(tiempos_en_sistema) if tiempos_en_sistema else 0

    return Ws_promedio
```

**Parametros**

```python
lam, mu, T = 4, 6, 1000
np.random.seed(42)
```

**De lo anterior, realizamos la prueba:**

```python
ws_simulaciones = [simulacion_ross_m_m_1(lam, mu, T) for _ in range(100)]
Ws_simulado = np.mean(ws_simulaciones)


Ws_analitico = 1 / (mu - lam)

print(f"Solución por Simulación: {Ws_simulado:.4f}")
```
